```python
def find_reflection_line(pattern):
    """
    Find the reflection line in the pattern. It returns the position of the line and its type ('vertical' or 'horizontal').
    If the pattern is symmetrical about a vertical line, the position is the index of the column just to the left of the line.
    If the pattern is symmetrical about a horizontal line, the position is the index of the row just above the line.
    """
    n_rows = len(pattern)
    n_cols = len(pattern[0])

    # Check for vertical reflection
    for col in range(n_cols // 2):
        if all(pattern[row][col] == pattern[row][n_cols - col - 1] for row in range(n_rows)):
            return col, 'vertical'

    # Check for horizontal reflection
    for row in range(n_rows // 2):
        if all(pattern[row][col] == pattern[n_rows - row - 1][col] for col in range(n_cols)):
            return row, 'horizontal'

    return None, None  # No reflection line found

def summarize_patterns(patterns):
    """
    Calculate the summary number for a list of patterns based on their reflection lines.
    """
    summary = 0
    for pattern in patterns:
        line_pos, line_type = find_reflection_line(pattern)

        if line_type == 'vertical':
            summary += line_pos
        elif line_type == 'horizontal':
            summary += 100 * line_pos

    return summary

# Example patterns from the puzzle
patterns = [
    ["#.##..##.", "..#.##.#.", "##......#", "##......#", "..#.##.#.", "..##..##.", "#.#.##.#."],
    ["#...##..#", "#....#..#", "..##..###", "#####.##.", "#####.##.", "..##..###", "#....#..#"]
]

# Calculate the summary
summary_number = summarize_patterns(patterns)
summary_number
```

```python
def is_vertical_reflection(pattern, col):
    for row in range(len(pattern)):
        if pattern[row][col] != pattern[row][-col-1]:
            return False
    return True

def is_horizontal_reflection(pattern, row):
    for col in range(len(pattern[0])):
        if pattern[row][col] != pattern[-row-1][col]:
            return False
    return True

def find_reflection_line(pattern):
    # Check for vertical reflection
    for col in range(len(pattern[0]) // 2):
        if is_vertical_reflection(pattern, col):
            return col, 'vertical'

    # Check for horizontal reflection
    for row in range(len(pattern) // 2):
        if is_horizontal_reflection(pattern, row):
            return row, 'horizontal'

    return None, None

def summarize_patterns(patterns):
    summary = 0
    for pattern in patterns:
        line_pos, line_type = find_reflection_line(pattern)
        if line_type == 'vertical':
            summary += line_pos
        elif line_type == 'horizontal':
            summary += 100 * line_pos

    return summary

def read_patterns_from_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read().strip()
    patterns = content.split('\n\n')
    return [pattern.split('\n') for pattern in patterns]

def main():
    file_path = 'input.txt'
    patterns = read_patterns_from_file(file_path)
    summary = summarize_patterns(patterns)
    print(f"Summary number: {summary}")

if __name__ == "__main__":
    main()
```
