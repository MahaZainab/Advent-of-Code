```python
# Function to determine the type and strength of each hand
def hand_rank(hand):
    values = "23456789TJQKA"
    counts = {v: hand.count(v) for v in values}
    sorted_counts = sorted(counts.values(), reverse=True)

    if sorted_counts == [4, 1]:  # Four of a kind
        hand_type = 7
    elif sorted_counts == [3, 2]:  # Full house
        hand_type = 6
    elif sorted_counts == [3, 1, 1]:  # Three of a kind
        hand_type = 5
    elif sorted_counts == [2, 2, 1]:  # Two pair
        hand_type = 4
    elif sorted_counts == [2, 1, 1, 1]:  # One pair
        hand_type = 3
    else:
        hand_type = 2  # High card

    # Determine the strength of the hand for tie-breaking
    strength = sum((14 - values.index(v)) * (4 ** count) for v, count in counts.items())
    return hand_type, strength

# Read data from the input file
def read_data(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        hands = {line.split()[0]: int(line.split()[1]) for line in lines}
    return hands

# Calculate the total winnings
def calculate_winnings(hands):
    # Sort the hands based on their rank and strength
    sorted_hands = sorted(hands.keys(), key=lambda hand: hand_rank(hand), reverse=True)

    # Calculate the winnings for each hand
    winnings = {hand: rank * hands[hand] for rank, hand in enumerate(sorted_hands, start=1)}
    return winnings, sum(winnings.values())

# Main function
def main():
    file_path = 'input.txt'
    hands = read_data(file_path)
    winnings, total = calculate_winnings(hands)

    # Output the results
    print("Winnings by Hand:")
    for hand, win in winnings.items():
        print(f"{hand}: {win}")

    print(f"\nTotal Winnings: {total}")

# Execute the main function
main()
```
